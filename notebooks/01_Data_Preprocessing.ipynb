Encoding</h2>

```python
df=df.copy()
df['Dep_time'] = df['Dep_hours']*60 + df['Dep_min']
df['Arrival_time'] = df['Arrival_hours']*60 + df['Arrival_min']

df['Duration'] = df['Duration_hours']*60 + df['Duration_min']

df.drop(['Dep_hours','Dep_min','Arrival_hours','Arrival_min','Duration_hours','Duration_min'], axis=1, inplace=True)


df['Full_Date'] = pd.to_datetime(dict(year=2019, month=df['Month'], day=df['Date']))
df['Day_of_week'] = df['Full_Date'].dt.dayofweek
df.drop(['Date','Full_Date'], axis=1, inplace=True)


df['Month_sin'] = np.sin(2 * np.pi * df['Month']/12)
df['Month_cos'] = np.cos(2 * np.pi * df['Month']/12)
df.drop('Month', axis=1, inplace=True)

df = pd.get_dummies(df, columns=['Airline', 'Source', 'Destination'],dtype=int, drop_first=True)


df
```

<h2 style="font-family: Poppins, sans-serif; font-size: 1.6rem; font-weight:600; color:rgba(150, 255, 255, 0.6); text-shadow: 0 0 6px rgba(56,189,248,0.8), 0 0 12px rgba(56,189,248,0.6);">
Check For Outliers</h2>

```python
df_after = df.copy()
numeric_cols = ['Price', 'Dep_time', 'Arrival_time', 'Duration']

for col in numeric_cols:
    q1 = df_after[col].quantile(0.25)
    q3 = df_after[col].quantile(0.75)
    iqr = q3 - q1
    lower_limit = q1 - (1.5 * iqr)
    upper_limit = q3 + (1.5 * iqr)
    
    df_after = df_after[(df_after[col] >= lower_limit) & (df_after[col] <= upper_limit)]

df_after.reset_index(drop=True, inplace=True)
df_after
```

```python
for col in numeric_cols:
    plt.figure(figsize=(10, 4))

    plt.subplot(1, 2, 1)
    plt.boxplot(df[col])
    plt.title(f"{col} - Before")

    plt.subplot(1, 2, 2)
    plt.boxplot(df_after[col])
    plt.title(f"{col} - After")

    plt.show()
```

<h2 style="font-family: Poppins, sans-serif; font-size: 1.6rem; font-weight:600; color:rgba(150, 255, 255, 0.6); text-shadow: 0 0 6px rgba(56,189,248,0.8), 0 0 12px rgba(56,189,248,0.6);">
EDA</h2>

```python
plt.figure(figsize=(30, 30))
sns.heatmap(df_after.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python

fig = make_subplots(
    rows=1, cols=3,
    subplot_titles=('Avg Price by Day of Week', 'Duration vs Price', 'Price by Stops')
)


avg_price_day = df_after.groupby("Day_of_week")["Price"].mean()
fig.add_trace(
    go.Scatter(x=avg_price_day.index, y=avg_price_day.values,
               mode='lines+markers', name='Avg Price'),
    row=1, col=1
)

fig.add_trace(
    go.Scatter(x=df_after["Duration"], y=df_after["Price"],
               mode='markers',
               marker=dict(color=df_after["Total_Stops"], colorscale='Viridis'),
               name='Duration vs Price'),
    row=1, col=2
)

avg_price_stops = df_after.groupby("Total_Stops")["Price"].mean()
fig.add_trace(
    go.Bar(x=avg_price_stops.index.astype(str), y=avg_price_stops.values, name='Avg Price per Stops'),
    row=1, col=3
)

fig.update_layout(height=500, width=1200, title_text="Flight Fare EDA Dashboard")
fig.show()
```
